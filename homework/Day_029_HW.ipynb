```
#library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math 
import statistics
import seaborn as sns
```

```
# Q1:計算標準常態分配，小於1的機率有多大? 
# Python實現常態分配
# 1.設定使用參數
mu=0
sigma=1
x = np.linspace(-1, 1, 500)
# 2.產生pdf圖
y = stats.norm.pdf(x,mu, sigma)
plt.plot(x, y, "b-", linewidth=2)
#plt.grid(True)
plt.title('pdf of Norma(0,1)')
plt.show()
# 3.計算常態分配的累積機率(cumulative density function),pdf的積分
cumsum_probs = stats.norm.cdf(x, mu, sigma)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('cdf of Normal(0,1)')
plt.plot(x, cumsum_probs,"b-", linewidth=2)
plt.show()
print(cumsum_probs[1])
```

```
# Q3:X~N(2,4),x 服從常態分配，平均數為2,變異數為 4，計算 X小於 6 的機率有多大?? 
mu=2
sigma=4
x = np.linspace(-6, 6, 500)
# 2.產生pdf圖
y = stats.norm.pdf(x,mu, sigma)
plt.plot(x, y, "b-", linewidth=2)
#plt.grid(True)
plt.title('pdf of Norma(0,1)')
plt.show()
# 3.計算常態分配的累積機率(cumulative density function),pdf的積分
cumsum_probs = stats.norm.cdf(x, mu, sigma)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('cdf of Normal(0,1)')
plt.plot(x, cumsum_probs,"b-", linewidth=2)
plt.show()
print(cumsum_probs[6])
```
