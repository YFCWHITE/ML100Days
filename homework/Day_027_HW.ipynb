```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
```

```
p = 0.5 # 事件A 機率 0.5
n = 100  # 重複實驗100次,
r = np.arange(0,100)

probs = stats.binom.pmf(r, n, p)
print('Probality=',probs[50])
plt.bar(r, probs)
plt.ylabel('P(X=x)')
plt.xlabel('x')
plt.title('binomial(n=100,p=0.5)')
plt.show()
```
