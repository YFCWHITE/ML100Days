```python
#1.畫出cos圖檔,並儲存
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)
```

```python
plt.plot(x ,y_cos)
#plt.show()
#將結果存成圖檔
plt.savefig("D18cos.png",dpi=300,format="png")
```

```python
#2.給定散點圖顏色
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.scatter(X,Y)
plt.show()
```
